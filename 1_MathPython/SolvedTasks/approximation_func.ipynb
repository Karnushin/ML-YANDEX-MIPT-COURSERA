```python
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5.) * np.exp(x/10.) + 5 * np.exp(-x/2.)
```

```python
def y1(x):
    a = np.array([[1,1], [1,15]])
    b = np.array([f(1), f(15)])
    w = solve(a,b)
    return w[0] + w[1] * x
```

```python
def y2(x):
    a = np.array([[1,1,1],[1,8,64], [1,15, 225]])
    b = np.array([f(1), f(8), f(15)])
    w = solve(a, b)
    return w[0] + w[1] * x + w[2] * x ** 2
```

```python
def y3(x):
    a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000], [1,15,225,15**3]])
    b = np.array([f(1), f(4), f(10), f(15)])
    w = solve(a, b)
    return w[0] + w[1] * x + w[2] * x ** 2 + w[3] * x ** 3
```

```python
#    x=[1,15],   = 0.1    
x = np.arange(1, 15, 0.1)
plt.figure(figsize=(15,10))
plt.plot(x, f(x), label = 'f(x), original')
plt.plot(x, y1(x), label = 'y1(x), 1st degree')
plt.plot(x, y2(x), label = 'y2(x), 2nd degree')
plt.plot(x, y3(x), label = 'y3(x), 3rd degree')
plt.legend(title = 'Graphics')
plt.show()
```
