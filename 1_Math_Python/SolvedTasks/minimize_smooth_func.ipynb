```python
import numpy as np
from scipy.optimize import minimize, differential_evolution
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5.) * np.exp(x/10.) + 5 * np.exp(-x/2.)
```

```python
def h(x):
    return f(x).astype(int)
```

```python
#построение графиков функций f(x) и h(x)=int(f(x))
plt.figure(figsize=(10,5))
x = np.arange(1,30,0.1)
plt.plot(x, h(x), label='int(f(x))')
plt.plot(x, f(x), label='f(x)')
plt.legend(title='Smooth and nosmooth')
plt.show()
```

## Минимизация гладкой функции


```python
#использование градиентного метода для решения задачи оптимизаци(нахождение минимума)
#легально: функция гладкая;убеждаемся, что нахождение лок. мин. зависит от нач. приближения
x_min2 = minimize(f, x0=2, method='BFGS')
print('x =', round(x_min2.x[0], 2), 'local min =', round(x_min2.fun,2))

x_min30 = minimize(f, x0=30, method='BFGS')
print('x =', round(x_min30.x[0], 2), 'local min =', round(x_min30.fun,2))
```

## Глобальная оптимизация

```python
x_minDE = differential_evolution(f, [(1,30)])
print('x =', round(x_minDE.x[0], 2), 'global min =', round(x_minDE.fun[0], 2))
```

##  Минимизация негладкой функции

```python
x_min_nosmooth = minimize(h, x0=30, method='BFGS')
print('x =', round(x_min_nosmooth.x[0], 2), 'local min =', round(x_min_nosmooth.fun[0],2))
x_minDE_nosmooth = differential_evolution(h, [(1,30)])
print('x =', round(x_minDE_nosmooth.x[0], 2), 'global min =', round(x_minDE_nosmooth.fun, 2))
```
